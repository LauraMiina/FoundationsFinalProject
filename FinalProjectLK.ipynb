```python
my_url = "https://climatereanalyzer.org/clim/t2_daily/json_cfsr/cfsr_world_t2_day.json"
raw_data = requests.get(my_url).json()

data_dict = {}
for entry in raw_data:
    year = entry['name']
    data = entry['data']
    data_dict[year] = data
df = pd.DataFrame(data_dict)
df
```

```python
df.to_csv('temperatures.csv', index=False)
```

```python
day_of_year = pd.Series(range(1, 367))
day_of_year = day_of_year[:len(df)]
df['Day_of_Year'] = day_of_year.values
df.set_index('Day_of_Year', inplace=True)
df.tail()
```

```python
import pandas as pd
import altair as alt

df = pd.DataFrame(data_dict)

day_of_year = pd.Series(range(1, len(df) + 1))
df['Day_of_Year'] = day_of_year
df.set_index('Day_of_Year', inplace=True)
df_melted = df.reset_index().melt(id_vars='Day_of_Year', var_name='Year', value_name='Value')
chart = alt.Chart(df_melted).mark_line().encode(
    x='Day_of_Year:Q',
    y='Value:Q',
    color='Year:N'
).properties(
    width=600,
    height=400,
    title='Value trends across years'
)
chart
```
